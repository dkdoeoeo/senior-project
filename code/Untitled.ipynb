```python
import re
from itertools import chain

with open('E:/專題/data/2022/scc game/scc20220101-4.txt', 'r', encoding='utf-8') as file:
    content = file.read()
    conlist = re.split(r'[<>]', content)
    conlist = [s for s in conlist if s]
    del conlist[1]
    newlist = []
    for item in conlist:
        sub = re.split(r'[ /]', item)
        for ele in sub:
            newlist.append(ele);
    newlist = [s for s in newlist if s]
    print(newlist)
```
